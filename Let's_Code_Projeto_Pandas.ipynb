## 2) Relatorio

```python
def relatorio_quantitativo(df, cols, font_size = 7):

    style = [
        dict(selector="th", props=[("font-size", f'{font_size + 1}pt')]),
        dict(selector="td", props=[("font-size", f'{font_size    }pt')])
    ]
        
    relatorio = df[cols].describe()
    mediana = pd.Series(df[cols].apply(np.nanmedian, axis = 0), name = 'mediana')
    relatorio = relatorio.append(mediana).loc[['min', '25%', 'mediana', '75%', 'max', 'mean']]

    display(relatorio.T.style.format('{:.2f}').set_table_styles(style))

relatorio_quantitativo(df, cols_num)
```

## 3) Relatório sem outliers

```python
def is_outlier(col):
    Q1, Q3 = col.quantile((0.25, 0.75))
    IQR = Q3 - Q1
    return (col > Q3 + 1.5 * IQR) | (col < Q1 - 1.5 * IQR)

def input_outlier(col, value = None, verbose = False):
    if isinstance(value, str):
        if value == 'mediana':
            value = np.nanmedian(col)
        else:
            raise ValueError('value deve ser numérico, None ou "mediana".')
    outliers = is_outlier(col)
    print(f'Inputando {outliers.sum()} outliers com o valor {value} ...')
    col[outliers] = value
    return col
```

```python
df_num_outlier_none = df[cols_num].apply(input_outlier)
```

```python
relatorio_quantitativo(df_num_outlier_none, cols_num)
```

## 4) Relatório com outliers tratados

```python
df_num_outlier_median = df[cols_num].apply(input_outlier, value = 'mediana')
```

```python
relatorio_quantitativo(df_num_outlier_median, cols_num)
```

## 5) Relatório das variáveis qualitativas

```python
def relatorio_qualitativo(df, cols, font_size = 7):

    style = [
        dict(selector="th", props=[("font-size", f'{font_size + 1}pt')]),
        dict(selector="td", props=[("font-size", f'{font_size    }pt')])
    ]
    
    for col in cols:
        rel = pd.DataFrame({'freq_abs': df[col].value_counts().sort_index(),
                            'freq_acc': df[col].value_counts().sort_index().cumsum(),
                            'freq_rel': df[col].value_counts(normalize = True).sort_index().map('{:.2f}'.format),
                            'freq_rac': df[col].value_counts(normalize = True).sort_index().cumsum().map('{:.2f}'.format),
                            })
        rel = pd.concat({col: rel}, names=['Feature'])
        display(rel.T.style.set_table_styles(style))


relatorio_qualitativo(df, cols_cat[1:])
```

## 7 e 8) Criando dummies e gerando relatório

```python
cols = cols_cat[2:5]
cols
```

```python
df_dummy = pd.get_dummies(df, columns = cols)
df_dummy = df_dummy.drop(cols_cat, axis = 1, errors = 'ignore')
df_dummy['ricos10'] = df_dummy[col_renda] > renda_alta
df_dummy.groupby('ricos10').agg('mean').style.format('{:.2f}').highlight_max()
```

```python
df_dummy = pd.get_dummies(df, columns = cols)
df_dummy = df_dummy.drop(cols_cat, axis = 1, errors = 'ignore')
df_dummy['faixa_renda'] = pd.cut(df_dummy[col_renda], df_dummy[col_renda].quantile(np.linspace(0, 1, 11)))
df_dummy.groupby('faixa_renda').agg('mean').T.style.format('{:.2f}').bar(axis = 1)
```
